```python
numeric_df=df.select_dtypes(include=[np.number])
non_numeric_df=df.select_dtypes(exclude=[np.number])
Q1=numeric_df.quantile(0.25)
Q3=numeric_df.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
numeric_df.shape
```

```python
non_numeric_df.shape
```

```python
numeric_df=numeric_df[~((numeric_df<(Q1-1.5*IQR)) | (numeric_df>(Q1+1.5*IQR)))]
numeric_df.shape
```

```python
numeric_df=numeric_df.fillna(numeric_df.mean())
```

```python
df=pd.concat([numeric_df,non_numeric_df],axis=1)
```

```python
df
```

```python
df.isnull().sum()
```

## Data Visualization

```python

sns.scatterplot(x=df['Player'],y=df['Ct'])
plt.title('Catches Taken by Player')
plt.xlabel('Player')
plt.ylabel('Catches')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.barplot(x=df['Player'],y=df['Dis'])
plt.title('Total Dismissals Made by the Player')
plt.xlabel('Player')
plt.ylabel('Dismissals')
plt.xticks(rotation=90)
plt.show()
```

```python
numeric_df=numeric_df.corr()
sns.heatmap(numeric_df,cmap='BrBG',annot=True)
plt.show()
numeric_df
```

```python
fig,ax=plt.subplots(figsize=(10,5))
sns.scatterplot(x=df['Player'],y=df['Inns'],ax=ax)
plt.title('Innings Played by Player')
plt.xlabel('Player')
plt.ylabel('Innings')
plt.xticks(rotation=90)
plt.show()
```
